```python
#This program calculates daily solar energy output (how much energy solar panels produce per day) in kWh
```

```python
def temperature_derate(temp_coeff, cell_temp):# Calculate temperature adjustment factor
    temp_derate = 1- (temp_coeff/100)*(cell_temp - 25)
    return temp_derate
def daily_energy_kwh(rated_power, num_panels, peak_sun_hours, system_efficiency, temp_coeff, cell_temp):
    temp_derate = temperature_derate(temp_coeff, cell_temp)#Get derate factor from temperature
    effective_power = rated_power * temp_derate #Adjust power of one panel
    array_power = effective_power * num_panels #Total power of all panels
    energy_wh = array_power * peak_sun_hours * system_efficiency #Daily energy in watt-hours
    energy_kwh=energy_wh/1000 #Convert to kilowatt-hours
    return energy_kwh
def main():
    #example
    rated_power=400
    num_panels=12
    peak_sun_hours=4.2
    system_efficiency=0.80
    temp_coeff=0.40
    cell_temp=35
    result = daily_energy_kwh(rated_power, num_panels, peak_sun_hours, system_efficiency, temp_coeff, cell_temp)
    print ('worked exp')
    print (f'Daily energy output: {result:.1f} kWh')

    #Test 1
    rated_power = 350 #Just changed numbers
    num_panels = 10
    peak_sun_hours = 5.0
    system_efficiency = 0.82
    temp_coeff = 0.35
    cell_temp = 20
    result = daily_energy_kwh(rated_power, num_panels, peak_sun_hours, system_efficiency, temp_coeff, cell_temp)
    print ("Test 1")
    print (f'Daily energy output: {result:.1f} kWh')

    #Test 2
    rated_power = 450
    num_panels = 8
    peak_sun_hours = 3.5
    system_efficiency = 0.78
    temp_coeff = 0.45
    cell_temp = 40
    result = daily_energy_kwh(rated_power, num_panels, peak_sun_hours, system_efficiency, temp_coeff, cell_temp)
    print ("Test 2")
    print (f'Daily energy output: {result:.1f} kWh')
    

main()

   
```
